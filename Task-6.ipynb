## Decision Tree Model Training

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
X_tain, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(X_tain,y_train)
```

```python
y_pred = dtc.predict(X_test)
y_pred
```

## Comapring Actual & Predicted values

```python
pd.DataFrame({"Actual":y_test, "Predicted":y_pred} )
```

## Visaulization of the trained model

```python
plt.figure(figsize=(16,12))
tree.plot_tree(dtc,
               class_names=iris.target_names,
               feature_names=iris.feature_names[:],
               filled=True,
               rounded=True,)

plt.show()
```

# <span style='color:red'>  **Feeding new data to this trained model** </span>

#### Predicting the class output for some random values Sepal length, Sepal width, Petal length & Petal length

```python
pre_result = dtc.predict([[5.8, 2.7, 5.1, 1.9]])
if pre_result == [0]:
    print("setosa")
elif pre_result == [1]:
    print("versicolor")
else:
    print("virginica")
```

 - <span style='color:green'> - here for the values  5.8, 2.7, 5.1, 1.9  our model predicted 'virginica' </span>

## Calculating model accuracy 

```python
import sklearn.metrics as sm
```

```python
print("Accuracy Score: ",sm.accuracy_score(y_test, y_pred))
```

- so the accuracy of our model is 0.97 or 96.66%
